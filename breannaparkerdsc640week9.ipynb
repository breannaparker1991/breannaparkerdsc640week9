```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

```python
#import data to dataframe, number 1
df1=pd.read_csv('costcos-geocoded.csv')
df1.head()
```

```python
#import data to dataframe, number 2
df2=pd.read_csv('ppg2008.csv')
df2.head()
```

```python
#heatmap of dataset1
sns.heatmap(df1.corr(), annot = True, cmap='Blues')
```

```python
#heatmap of dataset2
sns.heatmap(df2.corr(), annot = True, cmap='Blues')
```

```python
#getting the map
map = folium.Map(location=[df1.Latitude.mean(), df1.Longitude.mean()], zoom_start=34, control_scale=True)
```

```python
#mapping the data
for index, location in df1.iterrows():
    folium.Marker([location["Latitude"], location["Longitude"]], popup=location["City"]).add_to(map)
```

```python
#lollipop chart
fig, axes = plt.subplots() 
axes.stem(df2['G'], df2['MIN'], 
          use_line_collection=True, basefmt=' ')  
plt.xlabel('G') 
plt.ylabel('MIN') 
```
