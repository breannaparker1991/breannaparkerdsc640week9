```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.basemap import Basemap
import csv
import geopandas as gpd
from shapely.geometry import Point
import missingno as msn
```

```python
#import data to dataframe
df1=pd.read_csv('costcos-geocoded.csv')
df1.head()
```

```python
#import data to dataframe
df2=pd.read_csv('ppg2008.csv')
df2.head()
```

```python
#heatmap of dataset1
sns.heatmap(df1.corr(), annot = True, cmap='Blues')
```

```python
#heatmap of dataset2
sns.heatmap(df2.corr(), annot = True, cmap='Blues')
```

```python
#reimporting the data as a geodataframe
df = gpd.read_file(r'costcos-geocoded.csv')
df.head()
```

```python
#verifying the geodataframe was successful
type(df)
```

```python
#combining lat and lon values
df['coordinates'] = df[['Longitude', 'Latitude']].values.tolist()
df.head()
```

```python
#checking type and it's the wrong one
type(df.coordinates)
```

```python
# changing coordinates to geo POINT
df['coordinates'] = df['coordinates'].apply(Point)
df.head()
```

```python
# changing df to geodf
df = gpd.GeoDataFrame(df, geometry='coordinates')
df.head()
```

```python
#confirming that we now have geodata!
type(df.coordinates)
```

```python
df.plot()
```

```python
 # Plotting to see the hurricane overlay the US map:
fig, ax = plt.subplots(1, figsize=(30,20))
base = df.plot(ax=ax, color='#3B3C6E')

# plotting the hurricane position on top with red color to stand out:
df.plot(ax=base, color='darkred', marker="*", markersize=10);
```

```python
#lollipop chart
fig, axes = plt.subplots() 

axes.stem(df2['G'], df2['MIN'], 
          use_line_collection=True, basefmt=' ') 
  
# starting value of y-axis 
axes.set_ylim(0) 
  
# details and formatting of chart 
plt.title('') 
plt.xlabel('G') 
plt.ylabel('MIN') 
```
